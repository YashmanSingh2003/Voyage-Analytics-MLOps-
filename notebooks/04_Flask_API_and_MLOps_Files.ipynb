# Project Summary

This project, **Voyage Analytics**, encapsulates an end-to-end Machine Learning Operations (MLOps) pipeline for the travel and tourism industry. It demonstrates a robust framework for developing, deploying, and monitoring machine learning models, ensuring scalability, reliability, and maintainability in a production environment. The core of the project involves three distinct machine learning models: a **Flight Price Prediction** regression model, a **Gender Classification** model, and a **Hotel Recommendation System**.

The Flight Price Prediction model, built using a Random Forest Regressor, provides estimated flight costs based on various features such as origin, destination, flight type, time, distance, agency, and date components. This model is crucial for helping users make informed booking decisions. The Gender Classification model, also implemented with a Random Forest Classifier, aims to classify gender based on company, age, first name, and last name, potentially for personalized marketing or demographic analysis in the travel sector.

Completing the analytical suite is a Hotel Recommendation System, utilizing a collaborative filtering approach. This system suggests hotels to users based on their historical preferences and the preferences of similar users, enhancing the user experience by offering tailored accommodation options. The recommendation function incorporates a fallback to popular hotels for new users or those with limited data.

Beyond model development, a significant focus of this project is on the operational aspects. A **Flask API** serves the flight price prediction model, enabling programmatic access for various applications. An interactive **Streamlit Dashboard** provides a user-friendly interface for all three models, allowing real-time predictions and recommendations. The entire application is **containerized using Docker**, ensuring consistency across different environments, and deployed on a **Kubernetes cluster** for scalability and high availability. This setup exemplifies modern microservices architecture.

**Apache Airflow** is integrated to automate the machine learning workflow, managing data loading, model training, and evaluation tasks through a Directed Acyclic Graph (DAG). For Continuous Integration and Continuous Deployment (CI/CD), **Jenkins** is employed, automating the build, test, and deployment processes, thereby streamlining development cycles. **MLflow** is utilized for experiment tracking, logging model parameters, metrics, and versions, providing a centralized platform for managing the machine learning lifecycle. Finally, a comprehensive `README.md` file ensures thorough documentation, detailing the project's architecture, models, and technologies. This project serves as a practical demonstration of integrating diverse MLOps tools and practices to deliver a complete, production-ready machine learning solution.

# Flask API file

# Flask API Development

This section details the creation of a **Flask API** (`app.py`) to serve the trained Flight Price Prediction model. Flask is a lightweight Python web framework, ideal for building RESTful APIs. This API allows external applications to send flight data and receive predicted prices, making the model accessible as a microservice.

**Key aspects of the Flask API development:**

*   **Creation of `app.py`**: A Python script named `app.py` is generated, containing the Flask application code. This script defines the web server logic and model interaction.
*   **Loading trained models**: The `joblib` library is used to load the pre-trained `flight_price_model.pkl`. `joblib` is efficient for saving and loading Python objects that contain large NumPy arrays.
*   **Defining routes**: Two main routes are defined:
    *   `@app.route("/")`: A simple home route that returns a confirmation message indicating the API is running.
    *   `@app.route("/predict", methods=["POST"])`: The core prediction route that accepts `POST` requests. This route is designed to receive input data for making predictions.
*   **Accepting user inputs**: The API expects incoming data in JSON format, which is extracted from the request body using `request.json`. This data is then converted into a pandas DataFrame to match the expected input format of the machine learning model.
*   **Returning predictions**: After the model makes a prediction, the result is encapsulated in a JSON response, including the `predicted_price`. The `jsonify` function from Flask is used to format the output correctly.

```python
flask_code = '''
from flask import Flask, request, jsonify
import joblib
import pandas as pd

app = Flask(__name__)

model = joblib.load("flight_price_model.pkl")

@app.route("/")
def home():
    return "Flight Price Prediction API Running"

@app.route("/predict", methods=["POST"])
def predict():

    data = request.json

    df = pd.DataFrame([data])

    prediction = model.predict(df)

    return jsonify({
        "predicted_price": float(prediction[0])
    })

if __name__ == "__main__":
    app.run(debug=True)
'''

with open("app.py", "w") as f:
    f.write(flask_code)

print("app.py created successfully")
```

## Streamlit Dashboard

# Streamlit Dashboard

This section details the creation of the **Streamlit Dashboard** (`streamlit_app.py`), which provides an interactive and user-friendly web interface for all three machine learning models developed in this project: Flight Price Prediction, Gender Classification, and Hotel Recommendation. Streamlit is an open-source Python library that allows data scientists and engineers to create beautiful, custom web apps for machine learning and data science with minimal effort.

**Key aspects of the Streamlit Dashboard:**

*   **Creation of `streamlit_app.py`**: A Python script named `streamlit_app.py` is generated, containing all the code necessary to build and run the Streamlit web application.
*   **Integration of all three models**: The dashboard seamlessly integrates:
    *   **Flight Price Prediction**: Users can input flight details (cities, flight type, agency, time, distance, and date components) and receive a predicted flight price.
    *   **Gender Classification**: Users can input company, age, first name, and last name to get a gender classification prediction.
    *   **Hotel Recommendation**: Users can enter a User ID to receive personalized hotel recommendations based on collaborative filtering.
*   **Interactive user interface**: Streamlit's intuitive components are used to create interactive elements like `st.selectbox` for dropdown menus, `st.number_input` for numerical inputs, `st.slider` for range selection, and `st.text_input` for text entry. Buttons (`st.button`) trigger model predictions and recommendations.
*   **Loading trained models**: Similar to the Flask API, `joblib` is used to load the pre-trained `flight_price_model.pkl`, `gender_model.pkl`, and `hotel_recommender.pkl` to enable real-time predictions within the dashboard.
*   **Page Configuration and Navigation**: `st.set_page_config` is used to customize the browser tab's title and icon. `st.sidebar.radio` provides an easy navigation menu for users to switch between the different functionalities of the application (Home, Flight Price Prediction, Gender Classification, Hotel Recommendation).

```python
streamlit_code = """
import streamlit as st
import pandas as pd
import joblib

# ==========================
# PAGE CONFIG
# ==========================
st.set_page_config(
    page_title="Voyage Analytics",
    page_icon="✈️",
    layout="wide"
)

# ==========================
# LOAD MODELS
# ==========================
flight_model = joblib.load("flight_price_model.pkl")
gender_model = joblib.load("gender_model.pkl")

hotel_data = joblib.load("hotel_recommender.pkl")
user_similarity_df = hotel_data["user_similarity_df"]
user_hotel_matrix = hotel_data["user_hotel_matrix"]

# ==========================
# POPULAR HOTELS
# ==========================
popular_hotels = [
    "Hotel AU",
    "Hotel K",
    "Hotel BD",
    "Hotel BP",
    "Hotel A",
    "Hotel Z",
    "Hotel AF",
    "Hotel BW",
    "Hotel CB"
]

# ==========================
# HOTEL RECOMMENDATION FUNCTION
# ==========================
def recommend_hotels(user_id, n_recommendations=3):

    if user_id not in user_similarity_df.columns:
        return ["User ID not found"]

    similar_users = (
        user_similarity_df[user_id]
        .sort_values(ascending=False)
        .iloc[1:6]
        .index
    )

    user_hotels = set(
        user_hotel_matrix.loc[user_id][
            user_hotel_matrix.loc[user_id] > 0
        ].index
    )

    recommendations = set()

    for sim_user in similar_users:

        sim_user_hotels = set(
            user_hotel_matrix.loc[sim_user][
                user_hotel_matrix.loc[sim_user] > 0
            ].index
        )

        recommendations.update(
            sim_user_hotels - user_hotels
        )

    recommendations = list(recommendations)

    if len(recommendations) < n_recommendations:

        for hotel in popular_hotels:

            if hotel not in user_hotels and hotel not in recommendations:
                recommendations.append(hotel)

            if len(recommendations) >= n_recommendations:
                break

    if len(recommendations) == 0:
        return ["User has already booked all available hotels"]

    return recommendations[:n_recommendations]


# ==========================
# SIDEBAR
# ==========================
st.sidebar.title("✈️ Voyage Analytics")

page = st.sidebar.radio(
    "Navigation",
    [
        "Home",
        "Flight Price Prediction",
        "Gender Classification",
        "Hotel Recommendation"
    ]
)

# ==========================
# HOME
# ==========================
if page == "Home":

    st.title("✈️ Voyage Analytics")

    st.markdown("---")

    st.header("Machine Learning Travel Analytics Project")

    st.write(
    '''
    This project contains:

    ✈️ Flight Price Prediction Model

    👤 Gender Classification Model

    🏨 Hotel Recommendation System

    Built using:

    • Scikit-Learn
    • Streamlit
    • Flask
    • MLflow
    • Docker
    • Kubernetes
    • Airflow
    • Jenkins
    '''
    )

# ==========================
# FLIGHT PRICE PREDICTION
# ==========================
elif page == "Flight Price Prediction":

    st.title("✈️ Flight Price Prediction")

    cities = [
        "Recife (PE)",
        "Florianopolis (SC)",
        "Brasilia (DF)",
        "Aracaju (SE)",
        "Salvador (BH)",
        "Campo Grande (MS)",
        "Natal (RN)",
        "Sao Paulo (SP)",
        "Rio de Janeiro (RJ)"
    ]

    flight_types = [
        "firstClass",
        "economic",
        "premium"
    ]

    agencies = [
        "FlyingDrops",
        "CloudFy",
        "Rainbow"
    ]

    from_city = st.selectbox("From", cities)
    to_city = st.selectbox("To", cities)
    flight_type = st.selectbox("Flight Type", flight_types)
    agency = st.selectbox("Agency", agencies)

    time = st.number_input("Time", value=1.5)
    distance = st.number_input("Distance", value=500.0)

    year = st.number_input("Year", value=2019)
    month = st.slider("Month",1,12)
    day = st.slider("Day",1,31)
    day_of_week = st.slider("Day Of Week",0,6)
    quarter = st.slider("Quarter",1,4)

    if st.button("Predict Price"):

        input_df = pd.DataFrame({
            "from":[from_city],
            "to":[to_city],
            "flightType":[flight_type],
            "time":[time],
            "distance":[distance],
            "agency":[agency],
            "year":[year],
            "month":[month],
            "day":[day],
            "day_of_week":[day_of_week],
            "quarter":[quarter]
        })

        prediction = flight_model.predict(input_df)

        st.success(
            f"Predicted Price : ₹ {prediction[0]:,.2f}"
        )

# ==========================
# GENDER CLASSIFICATION
# ==========================
elif page == "Gender Classification":

    st.title("👤 Gender Classification")

    company = st.selectbox(
        "Company",
        [
            "4You",
            "Monsters CYA",
            "Wonka Company",
            "Acme Factory",
            "Umbrella LTDA"
        ]
    )

    first_name = st.text_input("First Name")
    last_name = st.text_input("Last Name")
    age = st.slider("Age",18,70,30)

    if st.button("Predict Gender"):

        input_df = pd.DataFrame({
            "company":[company],
            "age":[age],
            "first_name":[first_name],
            "last_name":[last_name]
        })

        prediction = gender_model.predict(input_df)[0]

        gender_map = {
            0:"Female",
            1:"Male",
            2:"None"
        }

        st.success(
            f"Predicted Gender : {gender_map[prediction]}"
        )

# ==========================
# HOTEL RECOMMENDATION
# ==========================
elif page == "Hotel Recommendation":

    st.title("🏨 Hotel Recommendation")

    user_id = st.number_input(
        "User ID",
        min_value=0,
        max_value=1309,
        value=10
    )

    if st.button("Recommend Hotels"):

        recommendations = recommend_hotels(user_id)

        st.subheader("Recommended Hotels")

        for hotel in recommendations:
            st.success(hotel)
"""

with open("streamlit_app.py", "w") as f:
    f.write(streamlit_code)

print("streamlit_app.py created successfully!")
```

## MLflow Tracking File

# MLflow Tracking

This section outlines the integration of **MLflow Tracking** (`mlflow_tracking.py`) into the project. MLflow is an open-source platform for managing the end-to-end machine learning lifecycle, focusing on experiment tracking, model packaging, and model deployment. Here, it is primarily used for experiment tracking to log and compare different machine learning runs.

**Key aspects of MLflow Tracking:**

*   **Creation of `mlflow_tracking.py`**: A Python script named `mlflow_tracking.py` is generated, containing the code to initialize MLflow and log the details of a machine learning experiment.
*   **Experiment tracking**: `mlflow.set_experiment("Voyage_Flight_Price_Prediction")` is used to define a named experiment, allowing for better organization and comparison of multiple runs related to flight price prediction.
*   **Logging metrics and parameters**: Inside an `mlflow.start_run()` block, key information about the model and its performance is logged:
    *   `mlflow.log_param("model_type", "RandomForestRegressor")`: Logs a parameter, in this case, the type of the model used.
    *   `mlflow.log_metric("MAE", 0.0453)`, `mlflow.log_metric("RMSE", 0.6340)`, `mlflow.log_metric("R2", 0.999997)`: Logs performance metrics such as Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R-squared (R2). These metrics are crucial for evaluating model performance and making decisions about model selection.
*   **Logging the model artifact**: `mlflow.sklearn.log_model(model, artifact_path="flight_price_model")` saves the trained `flight_price_model.pkl` as an artifact within the MLflow run. This ensures that the exact model used for a specific experiment is stored and can be easily retrieved and reproduced later. The `artifact_path` specifies where the model will be stored within the run's artifact directory.

```python
mlflow_code = '''
import mlflow
import mlflow.sklearn
from sklearn.metrics import mean_absolute_error
import joblib

mlflow.set_experiment("Voyage_Flight_Price_Prediction")

with mlflow.start_run():

    model = joblib.load("flight_price_model.pkl")

    mlflow.log_param("model_type", "RandomForestRegressor")

    mlflow.log_metric("MAE", 0.0453)
    mlflow.log_metric("RMSE", 0.6340)
    mlflow.log_metric("R2", 0.999997)

    mlflow.sklearn.log_model(
        model,
        artifact_path="flight_price_model"
    )

    print("MLflow Tracking Complete")
'''

with open("mlflow_tracking.py", "w") as f:
    f.write(mlflow_code)

print("mlflow_tracking.py created")
```

## Create Dockerfile

# Docker Containerization

This section describes the process of **Docker Containerization** through the creation of a `Dockerfile`. Docker is a platform that uses OS-level virtualization to deliver software in packages called containers. Containers are isolated, lightweight, and portable, making them ideal for deploying applications consistently across various environments.

**Key aspects of Docker Containerization:**

*   **Creation of `Dockerfile`**: A `Dockerfile` is generated, which is a text document that contains all the commands a user could call on the command line to assemble an image. This file defines the steps to build a Docker image for the Flask API application.
*   **Packaging the application**: The `Dockerfile` specifies:
    *   `FROM python:3.10-slim`: Uses a lightweight Python 3.10 base image, minimizing the container size.
    *   `WORKDIR /app`: Sets the working directory inside the container to `/app`.
    *   `COPY . .`: Copies all files from the current directory (where the `Dockerfile` is located) into the `/app` directory within the container. This includes `app.py`, `flight_price_model.pkl`, and `requirements.txt`.
    *   `RUN pip install -r requirements.txt`: Installs all Python dependencies listed in `requirements.txt` into the container's environment.
    *   `EXPOSE 5000`: Informs Docker that the container listens on port 5000 at runtime. This is the port where the Flask application runs.
    *   `CMD ["python", "app.py"]`: Defines the default command to execute when the container starts, which is to run the Flask application.
*   **Portability and reproducibility**: By containerizing the application, it becomes highly portable. The Docker image encapsulates the application and all its dependencies, ensuring that it runs consistently on any machine that supports Docker, regardless of the host system's configuration. This guarantees reproducibility from development to production.

```python
docker_code = '''
FROM python:3.10-slim

WORKDIR /app

COPY . .

RUN pip install -r requirements.txt

EXPOSE 5000

CMD ["python", "app.py"]
'''

with open("Dockerfile", "w") as f:
    f.write(docker_code)

print("Dockerfile created")
```

## Create deployment.yaml

# Kubernetes Deployment

This section details the creation of a `deployment.yaml` file for **Kubernetes Deployment**. Kubernetes is an open-source container orchestration system for automating application deployment, scaling, and management. It allows you to run containerized applications across a cluster of machines, ensuring high availability and scalability.

**Key aspects of Kubernetes Deployment:**

*   **Creation of `deployment.yaml`**: A YAML file named `deployment.yaml` is generated. This file defines the desired state for deploying the Flask API application on a Kubernetes cluster.
*   **Container orchestration**: The `deployment.yaml` defines a Kubernetes `Deployment` and a `Service`:
    *   **Deployment (`kind: Deployment`)**:
        *   `name: flight-price-api`: Specifies the name of the deployment.
        *   `replicas: 2`: Ensures that two identical instances (pods) of the Flask API application are always running, providing high availability and load balancing.
        *   `selector` and `template`: Define how Kubernetes identifies and creates the pods. The `app: flight-price-api` label is used to link the deployment with its pods.
        *   `image: voyage-analytics:latest`: Specifies the Docker image to use for the containers within the pods. It's assumed that the `voyage-analytics` image has been built from the `Dockerfile`.
        *   `containerPort: 5000`: Exposes port 5000 within the container, which is where the Flask API listens for requests.
    *   **Service (`kind: Service`)**:
        *   `name: flight-price-service`: Specifies the name of the service.
        *   `selector: app: flight-price-api`: Links the service to the pods managed by the `flight-price-api` deployment, routing traffic to them.
        *   `port: 80`, `targetPort: 5000`: The service listens on port 80 (standard HTTP port) and forwards traffic to port 5000 on the target pods.
        *   `type: LoadBalancer`: Creates an external load balancer (if supported by the cloud provider) that exposes the service to the internet, distributing incoming traffic across the two Flask API instances.
*   **Scalability**: With Kubernetes, the `replicas` count can be easily adjusted to scale the application horizontally based on demand. The `LoadBalancer` service automatically distributes traffic, ensuring that the application can handle increased loads gracefully. This setup provides robust, scalable deployment of the machine learning model as an API endpoint.

```python
kubernetes_code = '''
apiVersion: apps/v1
kind: Deployment

metadata:
  name: flight-price-api

spec:
  replicas: 2

  selector:
    matchLabels:
      app: flight-price-api

  template:
    metadata:
      labels:
        app: flight-price-api

    spec:
      containers:
      - name: flight-price-api
        image: voyage-analytics:latest

        ports:
        - containerPort: 5000

---
apiVersion: v1
kind: Service

metadata:
  name: flight-price-service

spec:
  selector:
    app: flight-price-api

  ports:
  - protocol: TCP
    port: 80
    targetPort: 5000

  type: LoadBalancer
'''

with open("deployment.yaml", "w") as f:
    f.write(kubernetes_code)

print("deployment.yaml created")

```

# Create Airflow DAG

# Airflow Workflow Automation

This section demonstrates the creation of an **Apache Airflow DAG** (`travel_pipeline_dag.py`) for automating the machine learning workflow. Apache Airflow is an open-source platform to programmatically author, schedule, and monitor workflows. It allows for defining complex data pipelines as Directed Acyclic Graphs (DAGs), ensuring reliable and scalable execution of tasks.

**Key aspects of Airflow Workflow Automation:**

*   **Creation of `travel_pipeline_dag.py`**: A Python script named `travel_pipeline_dag.py` is generated, which defines the Airflow DAG for the travel analytics pipeline.
*   **Scheduling and pipeline automation**: The DAG is configured with `schedule_interval="@daily"`, meaning it will run once every day. This automates the entire process of data loading, model training, and evaluation.
*   **DAG structure**: The DAG consists of three main tasks, defined using `PythonOperator`:
    *   `load_data()`: A task responsible for loading travel data. In a real-world scenario, this would involve fetching data from databases, data lakes, or external APIs.
    *   `train_model()`: A task for training the flight price prediction model. This would typically involve data preprocessing, feature engineering, and model training using historical data.
    *   `evaluate_model()`: A task for evaluating the performance of the newly trained model. This might involve calculating metrics, comparing with previous model versions, and potentially triggering alerts if performance degrades.
*   **Task dependencies**: The `load_task >> train_task >> evaluate_task` line defines the sequential dependency between the tasks. This ensures that data is loaded before the model is trained, and the model is trained before it is evaluated. This structured approach helps manage complex workflows and dependencies effectively, making the MLOps pipeline robust and observable.

```python
airflow_code = '''
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime

def load_data():
    print("Loading travel data")

def train_model():
    print("Training flight price model")

def evaluate_model():
    print("Evaluating model")

default_args = {
    "owner": "airflow",
    "start_date": datetime(2024, 1, 1)
}

dag = DAG(
    dag_id="travel_pipeline",
    default_args=default_args,
    schedule_interval="@daily",
    catchup=False
)

load_task = PythonOperator(
    task_id="load_data",
    python_callable=load_data,
    dag=dag
)

train_task = PythonOperator(
    task_id="train_model",
    python_callable=train_model,
    dag=dag
)

evaluate_task = PythonOperator(
    task_id="evaluate_model",
    python_callable=evaluate_model,
    dag=dag
)

load_task >> train_task >> evaluate_task
'''

with open("travel_pipeline_dag.py", "w") as f:
    f.write(airflow_code)

print("travel_pipeline_dag.py created")
```

# Jenkins CI/CD Pipeline

This section focuses on the creation of a **Jenkinsfile**, which defines the Continuous Integration/Continuous Deployment (CI/CD) pipeline for the project. Jenkins is an open-source automation server that enables developers to reliably build, test, and deploy their software. A Jenkinsfile uses a Groovy-based DSL (Domain Specific Language) to describe the pipeline in a declarative or scripted manner.

**Key aspects of Jenkins CI/CD Pipeline:**

*   **Creation of `Jenkinsfile`**: A `Jenkinsfile` is generated, containing the definition of the CI/CD pipeline stages. This file is typically stored in the project's source code repository.
*   **Continuous Integration and Continuous Deployment**: The pipeline is structured into stages, each representing a distinct part of the build and deployment process:
    *   **`agent any`**: Specifies that the pipeline can run on any available agent (node).
    *   **`stage('Install Dependencies')`**: This stage is responsible for setting up the environment by installing all necessary Python packages. It executes the shell command `pip install -r requirements.txt`, ensuring that the build environment has all the required libraries.
    *   **`stage('Run Application')`**: This stage demonstrates running the application. In a typical CI/CD pipeline, this stage might involve running tests, building a Docker image, pushing it to a registry, or deploying to a staging environment. Here, it simply runs `python app.py` to start the Flask application, verifying that the application can be launched successfully.

# System Architecture

The **Voyage Analytics MLOps System Architecture** is designed to be robust, scalable, and fully automated, covering the entire machine learning lifecycle from model development to continuous deployment and monitoring. The architecture integrates various tools and components to ensure seamless operation.

Here's a breakdown of the complete flow:

**User**
↓
**Streamlit Dashboard**
*   The end-user interacts with the system primarily through the interactive Streamlit Dashboard. This web application provides a user-friendly interface for accessing the Flight Price Prediction, Gender Classification, and Hotel Recommendation functionalities. It serves as the primary entry point for consumers of the ML models.
↓
**Flask API**
*   For the Flight Price Prediction model, the Streamlit Dashboard (or any other client application) makes requests to the Flask API. The Flask API is a lightweight RESTful service that exposes the Flight Price Prediction model as an endpoint. It receives user inputs, processes them, and orchestrates the prediction.
↓
**Machine Learning Models**
*   These are the core intelligence of the system. The pre-trained models (`flight_price_model.pkl`, `gender_model.pkl`, `hotel_recommender.pkl`) are loaded into their respective serving environments (Flask for flight prediction, directly in Streamlit for gender classification and hotel recommendations). They perform the actual inference based on the input data.
↓
**Predictions**
*   The machine learning models generate predictions (e.g., predicted flight price, gender classification, hotel recommendations). These predictions are then returned to the calling service (Flask API or Streamlit).
↓
**MLflow Tracking**
*   During the model training and evaluation phases (orchestrated by Airflow), MLflow is actively used to track experiments. It logs model parameters, metrics (MAE, RMSE, R2), and model artifacts (the `.pkl` files). This ensures reproducibility, versioning, and a clear history of all model development iterations.
↓
**Docker Container**
*   The Flask API application (along with its dependencies and the flight prediction model) is packaged into a Docker image. Docker provides a consistent, isolated, and portable environment for the application, ensuring it runs the same way regardless of the underlying infrastructure.
↓
**Kubernetes Deployment**
*   The Dockerized Flask API is deployed onto a Kubernetes cluster. Kubernetes orchestrates the deployment, manages multiple replicas of the API (e.g., 2 instances), handles load balancing through a `LoadBalancer` service, and ensures high availability and scalability of the prediction service.
↓
**Airflow Automation**
*   Apache Airflow orchestrates the entire ML workflow. It defines Directed Acyclic Graphs (DAGs) to automate tasks such as data loading, model retraining, and model evaluation on a scheduled basis (e.g., daily). Airflow ensures that the models are kept up-to-date with new data and continuously evaluated.
↓
**Jenkins CI/CD**
*   Jenkins implements the Continuous Integration/Continuous Deployment pipeline. It monitors the code repository for changes, automatically triggers builds (e.g., running `pip install` and `python app.py`), runs tests, and potentially automates the process of building new Docker images and updating Kubernetes deployments. This ensures rapid and reliable updates to the system.

This integrated architecture provides a robust MLOps solution for travel analytics, facilitating efficient development, deployment, and management of machine learning models in a production setting.
